```python
import pandas as pd
import tensorflow_hub as hub
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv('all_data.csv')
human_texts = data[data.labels == 0]['text'].values
ai_texts = data[data.labels == 1]['text'].values

embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

human_embeddings = embed(human_texts)
ai_embeddings = embed(ai_texts)

similarities = cosine_similarity(human_embeddings, ai_embeddings)

print(f"Average similarity: {np.mean(similarities):.3f}")

most_similar = np.argsort(-similarities.ravel())[:10] # top 10 most similar pairs
for idx in most_similar:
    human_idx, ai_idx = np.unravel_index(idx, similarities.shape)
    human_text = human_texts[human_idx]
    ai_text = ai_texts[ai_idx]
    similarity = similarities[human_idx, ai_idx]
    print(f"Similarity: {similarity:.3f}")
    print(f"Human text: {human_text}")
    print(f"AI-generated text: {ai_text}")
    print("-" * 50)
```
